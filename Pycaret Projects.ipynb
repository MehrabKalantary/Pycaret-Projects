```python
from pycaret import regression, classification, clustering
import pandas as pd
from pycaret.datasets import get_data
```

## Regression Problem

```python
data = get_data('boston')
```

```python
regression.setup(data=data, target='medv', session_id=12)
```

```python
best = regression.compare_models()
```

```python
best
```

```python
regression.plot_model(best)
```

```python
regression.evaluate_model(best)
```

---
## Classification Problem

```python
data = get_data('juice')
```

```python
classification.setup(data, target='Purchase', session_id=10)
```

```python
best = classification.compare_models()
```

```python
best
```

```python
lda = classification.create_model('lda')
```

```python
classification.plot_model(lda)
```

```python
classification.evaluate_model(lda)
```

## Clustering Problem

```python
data = get_data('jewellery')
```

```python
clustering.setup(data, session_id=15)
```

```python
clustering.models()
```

```python
model = clustering.create_model('kmeans')
```

```python
clustering.plot_model(model)
```

```python
clustering.evaluate_model(model)
```

```python
model2 = clustering.create_model('hclust')
```

```python
clustering.evaluate_model(model2)
```
